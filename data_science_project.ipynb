## Question 1 - Does the average salary change based on gender?

To address this question we pull out the columns on Gender and Salary (also Respondent as an index to take a count) clean up the data so that we have small number of categories that can compared. Since there is still information in the individuals who do not choose to report thier gender, we will retain these individuals as its own category. However we will need to remove individuals from the analysis who do not report salary as they are not informative. we will compare average salary (discrete values) using a bar plot. and will also look at distributions of thier salary (continuous data) using histograms.  

```python
q1_pd = pd_1.loc[:, ["Respondent", "Gender", "Salary"]]
```

### filling NAs

When a gender is missing I fill the missing value with x - This categiory will be retained and included in analysis. When Salary is missing I fill it with -1. the goal will be to remove but first I would like to retain information use the function 'fillna'

```python
q1_pd["Gender"] = q1_pd["Gender"].fillna("x")
q1_pd["Salary"] = q1_pd["Salary"].fillna(-1)
```

### group rows by value of Gender and aggregate

```python
q1_result_pd = q1_pd\
    .groupby(["Gender"])\
    .agg({"Respondent": "count", "Salary": "mean"})\
    .reset_index()\
    .rename(columns = {"Respondent": "Count", "Salary": "Average_Salary"})
```

### remove all the empty salaries

 in this analysis I want to remove all the empty salaries. These datapoints will not add to the analysis on a study of salary differences across genders. I am only interested in retaining individuals who reported salaries

```python
q1_result_pd.query("Average_Salary != -1")# removing -1s usinging sql; q1_result_pd[q1_result_pd["Average_Salary"] != -1] panads syntax for removing -1
```

### cleaning up the data into 4 categories 

```python
q1_pd_new = pd_1.loc[:, ["Respondent", "Gender", "Salary"]]
```

```python
def clean_gender(gender):
    """combines all small count gender groups into one large category OTHER
    Args:
        gender
    Returns:
        the given gender if it is listed as either Male female or not listed OR assigns it to OTHER
    """
    keep = {"Male", "Female", "x"}
    if gender in keep:
        return gender
    return "Other"
```

 apply the function to the data to combine all small gender categories into one group called 'other'

```python
q1_pd_new["Gender"] = q1_pd_new["Gender"].fillna("x").apply(clean_gender)
q1_pd_new["Salary"] = q1_pd_new["Salary"].fillna(-1)
```

## Question 3 - Does company size correlate to job satisfaction? 

To address this question I pulled out columns describing company size and job satisfaction. Missing data is droped for this analysis since it does not contribute to addressing the question. we need to convert the catergorical data into numerical indicators to perform the correlation. we will let apply 2 differnt models (linear regression and logistic regression) to fit the data and look at the coefficient of determination to determine if company size is correlates to job satisfaction. 

```python
q3_df = pd_1[["CompanySize", "JobSatisfaction"]]
```

### drop the missing fields

 for this analysis I droped individuals with missing values since it would not be informative for the analysis
 I are looking for the correlation between two features thus those features need to be reported on for analysis
 I used dropna function from pandas

```python
q3_df=q3_df.dropna()
```

### convert the categorical data into indicator (numerical) dummies using get_dummies

```python
size_dummies = pd.get_dummies(q3_df.CompanySize, prefix="sz")
```

```python
data = pd.concat([q3_df, size_dummies], axis=1)
```

```python
features = list(size_dummies.columns)
```

```python
data.head()
```

### from sklearn import logistic regression model and linear regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
X = data[features].values
y = data["JobSatisfaction"]

print(y)
model = LogisticRegression(max_iter=5000)
model.fit(X, y)
```

```python
print(features, model.coef_)

```

```python
logisticReg = model.score(X, y)
# r-squared - coefficient of determination 
# low = size of company is very poor at predicting job satisfaction
```

```python
logisticReg
```

```python
len(X)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = data[features].values
y = data["JobSatisfaction"]

print(y)
model2 = LinearRegression()
model2.fit(X, y)
```

```python
print(features, model2.coef_)
```

```python
LinearReg = model2.score(X, y)
```

```python
LinearReg
```

 based on these two models, and thier r2 values being low amd close to zero,
 job satisfaction is not correlated with company size. These r-squared are put into table form to make it easier to compare
